## Problem 1:
In general, for GPR with some kernel $k(x, x^{'})$, our posterior predictive mean is <br>
$\mu_{*}(x)=m(x)+k(\textbf{x}, x)^{T}(k(\textbf{x},\textbf{x})+\sigma^2 I)^{-1}(\bar{y}-m(\textbf{x})),$<br> where $\textbf{x}$ is our training x values, $x$ is our testing x values, and $\bar{y}$ is our training y values. The covariance matrix is <br>
$\Sigma_{*} = \left(k(x,x)+\sigma^2\right)-k(\textbf{x},x)^{T}\left(k(\textbf{x},\textbf{x})+\sigma^2I\right)^{-1}k(\textbf{x},x).$

### 1(a):
Suppose we have the zero mean function and our data is noise-free. Then<br> $\mu_{*}(x)=k(\textbf{x}, x)^{T}k(\textbf{x},\textbf{x})^{-1}\bar{y}$.<br> We see that this posterior predictive mean goes through our training dataset by plugging in $x=\textbf{x}$ (testing on the training data).
\begin{align}
\mu_{*}(\textbf{x}) &= k(\textbf{x}, \textbf{x})^{T}k(\textbf{x},\textbf{x})^{-1}\bar{y}\\
\end{align}
Let $x=(x_{1}, x_{2},...,x_{n})$. Then,
\begin{align}
k(\textbf{x}, \textbf{x})&=
\begin{bmatrix}
k(x_{1}, x_{1}) & k(x_{1}, x_{2}) & \dots & k(x_{1}, x_{n}) \\
k(x_{2}, x_{1}) & k(x_{2}, x_{2}) & \dots & k(x_{2}, x_{n}) \\
\vdots & \ddots & & \vdots \\
k(x_{n}, x_{1}) & k(x_{n}, x_{2}) & \dots & k(x_{n}, x_{n}) \\
\end{bmatrix}.
\end{align}
A propert of kernel functions $k(x_{i}, x_{j})$ is that $k(x_{i}, x_{j})= k(x_{j}, x_{i})$. Then we can see that $k(\textbf{x},\textbf{x})$ is symmetric, so $k(\textbf{x},\textbf{x})^{T}=k(\textbf{x},\textbf{x})$. So,
\begin{align}
\mu_{*}(\textbf{x}) &= k(\textbf{x}, \textbf{x})^{T}k(\textbf{x},\textbf{x})^{-1}\bar{y}\\
 &= k(\textbf{x}, \textbf{x})k(\textbf{x},\textbf{x})^{-1}\bar{y}\\
 &= I\bar{y}\\
 &= \bar{y}.
\end{align}
In other words, with the zero mean function and no noise, the mean of the posterior predictive distribution goes through the training data.<br>
When evaluated on the training data, our covariance matrix becomes
\begin{align}
\Sigma_{*}&=k(\textbf{x},\textbf{x})-k(\textbf{x},\textbf{x})^{T}k(\textbf{x},\textbf{x})^{-1}k(\textbf{x},\textbf{x})\\
&=k(\textbf{x},\textbf{x})-k(\textbf{x},\textbf{x})k(\textbf{x},\textbf{x})^{-1}k(\textbf{x},\textbf{x})\\
&=k(\textbf{x},\textbf{x})-Ik(\textbf{x},\textbf{x})\\
&=k(\textbf{x},\textbf{x})-k(\textbf{x},\textbf{x})\\
&= 0.
\end{align}
Therefore, at any training point, the covariance is 0 with the zero mean funciton and no noise.

### 1(b):
Now assume we have a non-zero mean function $m(x)$ but that we still have no noise. When evaluating on our training data ($x=\textbf{x}$), we have
\begin{align}
\mu_{*}&=m(\textbf{x})+k(\textbf{x},\textbf{x})^{T}k(\textbf{x},\textbf{x})^{-1}(\bar{y}-m(\textbf{x}))\\
&=m(\textbf{x})+k(\textbf{x},\textbf{x})k(\textbf{x},\textbf{x})^{-1}(\bar{y}-m(\textbf{x}))\\
&=m(\textbf{x})+I(\bar{y}-m(\textbf{x}))\\
&=\bar{y}.
\end{align}
Therefore, with no noise, the mean of the posterior predictive distribution goes through the training data even with a non-zero mean function. <br><br>
The covariance matrix does not depend on the mean function. Therefore, with no noise and a non-zero mean function, the covariance at any training point is still $0$.

### 1(c): 
Now assume we have the zero mean function once more, but this time we add postive noise $\sigma^2$. Then the mean of our posterior predictive distribution evaluated on the training data becomes
\begin{align}
\mu_{*} &= k(\textbf{x},\textbf{x})^{T}(k(\textbf{x},\textbf{x})^{T}+\sigma^2I)^{-1}\bar{y}\\
&= k(\textbf{x},\textbf{x})[k(\textbf{x},\textbf{x})+\sigma^2I]^{-1}\bar{y}.
\end{align}
Since $\sigma^2 >0$, $ k(\textbf{x},\textbf{x})+\sigma^2I\neq  k(\textbf{x},\textbf{x})$, so $\mu_{*}\neq \bar{y}$ . Thus, when we add noise, the posterior prediction mean no longer goes through the training data.

Now consider the covariance evaluated on the training data. This becomes
\begin{align}
\Sigma_{*}&=(k(\textbf{x},\textbf{x})+\sigma^2I)-k(\textbf{x},\textbf{x})^{T}(k(\textbf{x},\textbf{x})+\sigma^2I)^{-1}k(\textbf{x},\textbf{x}).
\end{align}
Unlike in part a, this is clearly not equal to $0$. So when we add positive noise, the variance of the posterior predictive distribution becomes non-zero.

### 1(d):
Bayesian Linear Regression is related to a GPR with kernel function
\begin{align}
k(x,x^{\prime})&=\phi(x)^{T}S_{0}\phi(x^{\prime})
\end{align}
where the true function generating the data is $f(x;w)=w^{T}\phi(x)$ and the prior is $p(w)=N(0, S_{0})$. Even if we assume no noise, the mean of the posterior predictive distribution obtained using Bayesian Linear Regression does not go through the training data like GPR with no noise despite the equivalency. This is not a contradiction because, unlike Bayesian Linear Regression, the results obtained in the above parts of the problem assumed no prior knowledge of the weights. Evaluating Murphy 17.45 over the training data and assuming no noise shows that the mean of the posterior predictive distribution is 
\begin{align}
\mu_{*}&=[\phi(\textbf{x})S_{0}\phi(\textbf{x})]^{T}[\phi(\textbf{x})S_{0}\phi(\textbf{x})^{T}]^{-1}\bar{y}.
\end{align}
This does not simplify to $\bar{y}$, so adding prior information about the true model weights changes the results of the equivalent Gaussian Process.

### 1(e):
Consider the simple case where we have the zero mean function and no noise. Suppose we scale the kernel function by a constant $c$. Then the mean of the posterior predictive distribution is 
\begin{align}
\mu_{*}&=[ck(\textbf{x}, \textbf{x})]^{T}[ck(\textbf{x}, \textbf{x})]^{-1}\bar{y}
\end{align}
When evaluated on the training data.
Since we are simply multiplying each entry by $c$, $ck(\textbf{x}, \textbf{x})$ is still symetric, so this simplifies to $\bar{y}$.
Now consider the variance evaluated on the training data.
\begin{align}
\Sigma_{*}&=ck(\textbf{x},\textbf{x})-[ck(\textbf{x},\textbf{x})]^{T}[ck(\textbf{x},\textbf{x})]^{-1}[ck(\textbf{x},\textbf{x})]\\
&=[ck(\textbf{x},\textbf{x})]-[ck(\textbf{x},\textbf{x})][ck(\textbf{x},\textbf{x})]^{-1}[ck(\textbf{x},\textbf{x})]\\
&=[ck(\textbf{x},\textbf{x})]-I[ck(\textbf{x},\textbf{x})]\\
&=[ck(\textbf{x},\textbf{x})]-[ck(\textbf{x},\textbf{x})]\\
&= 0.
\end{align}
Therefore, scaling the kernel function by $c$ does not affect the posterior predictive distribution when there is no noise and a zero mean function.

Now suppose there is some non zero mean function $m(x)$. Repeating the calculations from part b reveals that the posterior predictive distribution is once more unaffected since the variance does not depend on the mean function and the transposed scaled kernel times the inverse scaled kernel still results in the identity matrix.

Finally, suppose we have the zero mean function but we also have noise present in our data. From part c, we see that the mean of the posterior predictive distribution evaluated on the training data is
\begin{align}
\mu_{*} &= ck(\textbf{x},\textbf{x})^{T}([ck(\textbf{x},\textbf{x})]^{T}+\sigma^2I)^{-1}\bar{y}\\
&= ck(\textbf{x},\textbf{x})[ck(\textbf{x},\textbf{x})+\sigma^2I]^{-1}\bar{y}.
\end{align}
Once more, we see that this does not go through the training data. Furthermore, the scalar $c$ scaling the kernel function is also present in the inverse, so the mean of the posterior predictive distribution is not affected. Similarly, the variance of the distribution is not affected. <br><br>
Therefore, scaling the kernel function by a scalar $c$ does not affect the posterior predictive distribution.

### 1(f):
Suppose we implement GPR oin data where two observations have the same x value. One problem that will be encountered is that the GPR is only able to make one prediction at that particular x value. Let $x_{1}=x_{2}$. Then the covariance between $x_{1}$ and $x_{2}$ is $0$, in other words, $k(x_1,x_2) = k(x_2,x_1)=0$. Since GPR predictions are based on the covariance matrix, adding a zero to each row will remove a dimension from the predictions. Therefore, GPR loses a dimension in its predictions when our data has two observations at the same x-value.

## Problem 2:

We are given the NLL and the partial derivatives required to implement gradinet descent. The code below implements those operations.

```python
# imports
import csv
import numpy as np
import pandas as pd
from scipy import linalg
import seaborn as sns
import matplotlib.pyplot as plt
import random as rand
import math
import scipy.stats
from scipy.stats import gaussian_kde
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D
from pprint import pprint
```

```python
# convert the provided matlab code to python
def kernel(xi, xj, s):
    return np.exp(-0.5 / (s**2) * (xi - xj) ** 2)

def NLL(x_vec, y_vec, s_val, sig2_val):
    # n = x_vec.shape[0]
    n = len(x_vec)
    xx1, xx2 = np.meshgrid(x_vec, x_vec) # sneaky helper to make the K function easier to compute
    C = kernel(xx1, xx2, s_val) + sig2_val * np.eye(n)
    
    # sum up 3 terms
    lp = (1/2) * (y_vec.T @ np.linalg.solve(C, y_vec) + np.log(np.linalg.det(C)) + n * np.log(2 * np.pi))
    return lp

def grad_NLL(x_vec, y_vec, s_val, sig2_val):
    n = x_vec.shape[0]
    xx1, xx2 = np.meshgrid(x_vec, x_vec) # sneaky helper to make the K function easier to compute
    C = kernel(xx1, xx2, s_val) + sig2_val * np.eye(n)
    
    # dC/ds and dC/dsigma2
    gC_s = ((xx2 - xx1) ** 2) * C / s_val**3
    gC_sigma = np.eye(n)
    alpha = np.linalg.solve(C, y_vec)
    
    # apply formula for s term
    g_s = (1 / 2.0) * (np.trace(np.linalg.solve(C, gC_s)) - alpha.T @ gC_s @ alpha)
    # apply formula for sigma^2 term
    g_sigma2 = (1 / 2.0) * (np.trace(np.linalg.solve(C, gC_sigma)) - alpha.T @ gC_sigma @ alpha)
    
    return np.array([g_s, g_sigma2])
```

### 2(a):
Using the zero mean function and a Gaussian kernel, we implement GPR to the training data. We also use bandwidth $s=1$ and noise variance $\sigma^2=1$.

```python
# read in data
x = np.loadtxt('hw6_q2_train.csv', delimiter=',', usecols=(0), unpack=True, dtype=float)
y = np.loadtxt('hw6_q2_train.csv', delimiter=',', usecols=(1), unpack=True, dtype=float)
```

```python
s = 1
sigma2 = 1
n = len(x)
K = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        K[i, j] = np.exp(-0.5 / (s ** 2) * (x[i] - x[j]) ** 2)
                

C = K + sigma2 * np.eye(n)
model = np.linalg.solve(C, y)
```

```python
# define a function that makes prediction
def predict(xi, s):
    # initiate an empty list to store prediciton values
    yy = []

    for xval in xi:
        # calculate the kernel vector
        vec = []
        for i in range(len(x)):
            vec.append((x[i] - xval) ** 2)
        vec = np.array(vec)
        kvec = np.exp(-0.5 / (s ** 2) * vec)
        # make prediction
        yval = np.dot(kvec.T, model)
        yy.append(yval)

    return yy
```

```python
# write a function to return the standard deviation
def standard_deviation(xi,s,sigma2):
    # initiate an empty list to store prediciton values
    vari = []
    # calculate the squared two norm
    for xval in xi:
        # calculate the squared two norm
        vec = []
        for i in range(len(x)):
            vec.append((x[i] - xval) ** 2)
        vec = np.array(vec)
        kvec = np.exp(-0.5 / (s ** 2) * vec)
        # get variance
        vari.append(sigma2 + 1 - np.dot(kvec, model))
    return np.sqrt(vari)
```

```python
# plot training observations
plt.scatter(x, y, label='data')
# make predictions for the domain of the plot
mean_preds = predict(np.sort(x), s)
plt.plot(np.sort(x), mean_preds, color='red', label='mean fit')
# plot the standard deviation
sd = standard_deviation(mean_preds,s, sigma2)
plt.plot(np.sort(x), mean_preds + sd, color='orange', label='mean fit +- sd')
plt.plot(np.sort(x), mean_preds - sd, color='orange')
# format
plt.title("GPR with 0 mean function, s=$\sigma^2$=1")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

### 2(b):
We now implement gradient descent to find the $s$ and $\sigma^2$ values that minimize the NLL of the training set. To do so, we pick an initial state and update by subtraciting the learning rate times the gradient of the NLL. We continue updating the current state in this manner until satisfied. We are satisfied with our descent when our current state, $\begin{bmatrix} s\\ \sigma^2\end{bmatrix}$, changes by no more than $\begin{bmatrix} 0.01\\ 0.01\end{bmatrix}$.

```python
# define the initial state
s_list = [0.01 * i for i in range(1,201)]
sigma2_list = [0.01 * i for i in range(1, 201)]
state_list = [np.array([s, sigma2]) for s, sigma2 in zip(s_list, sigma2_list)]
# pick learning rate 
eta_list = [0.1 * i for i in range(1, 11)]
```

```python
# initiate lists to store states and NLL values
states = []
NLL_vals = []
cutoff = np.array([0.01, 0.01])
for state in state_list:
    for eta in eta_list:
        # define condition to terminate descent
        cond = True 
        new = np.copy(state)
        max_iters = 500
        iteration = 1
        while cond: # and iteration < max_iters:
            step = eta * grad_NLL(x, y, new[0], new[1])
            # we are only interested in positive values
            new[0] = max(0.001, new[0] - step[0])
            new[1] = max(0.001, new[1] - step[1])
            cond = np.all((np.abs(step) / eta) > cutoff)
            iteration += 1
        states.append(new)
        nll = NLL(x,y,new[0],new[1])
        NLL_vals.append(nll)
    
```

```python
index = np.argmin(np.array(NLL_vals))
nll = NLL_vals[index]
s = states[index][0]
sig2 = states[index][1]
[s, sig2, nll]
```

Using a variety of different starting points and step sizes, the optimal values of $s$ and $\sigma^2$ are

\end{align}
